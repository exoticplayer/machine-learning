<h3 id="最大似然估计-(MLE)">最大似然估计 (MLE)<a class="anchor-link" href="#最大似然估计-(MLE)">¶</a></h3><ul>
<li>给定随机样本 $\mathbf{X}=\left\{ \mathbf{x}_1,\mathbf{x}_2,...,\mathbf{x}_N \right\} =\left\{ \mathbf{x}_k \right\} _{k=1}^{N}$来自概率密度 $p\left( \mathbf{x}\mid \theta \right)$ </li>
<li>假设样本是独立同分布的,则它们的联合概率分布为 $p\left( \mathbf{X}\mid \theta \right) =p\left( \mathbf{x}_1,\mathbf{x}_2,...,\mathbf{x}_N\mid \theta \right) =\prod_{k=1}^N{p}\left( \mathbf{x}_k\mid \theta \right)$</li>
<li>估计使似然函数取最大值的参数 $\hat{\theta}:\hat{\theta}=\underset{\theta}{\text{arg}\max}\prod_{k=1}^N{p}\left( \mathbf{x}_k\mid \theta \right)$ </li>
<li>令似然函数对$\theta$的偏导数为零,求解 $\hat{\theta}:\frac{\partial}{\partial \theta}\log \prod_{k=1}^N{p}\left( \mathbf{x}_k\mid \theta \right) =0$</li>
<li>定义对数似然函数：  $L(\theta)=\log \prod_{k=1}^{N} p\left(\mathbf{x}_{k} \mid \theta\right) $</li>
<li>令似然函数对  $\theta$  的偏导数为零: 
$$
\begin{array}{l}
0&amp;=\frac{\partial}{\partial \theta} L(\theta) \\
&amp;=\frac{\partial}{\partial \theta} \log \prod_{k=1}^{N} p\left(\mathbf{x}_{k} \mid \theta\right) \\
&amp;=\sum_{k=1}^{N} \frac{\partial}{\partial \theta} \log p\left(\mathbf{x}_{k} \mid \theta\right) \\
&amp;=\sum_{k=1}^{N} \frac{1}{p\left(\mathbf{x}_{k} \mid \theta\right)} \frac{\partial}{\partial \theta} p\left(\mathbf{x}_{k} \mid \theta\right)
\end{array}
$$</li>
<li>求得  $\hat{\theta}$ , 对于样本点进行预测  $y=p(x \mid \hat{\theta})$ </li>
</ul>



<p>例：假设数据集 $ \mathbf{X}=\left\{\mathbf{x}_{1}, \mathbf{x}_{2}, \ldots, \mathbf{x}_{N}\right\}$ , 服从高斯分布 $ \mathcal{N}\left(\mu, \sigma^{2}\right)$ , 其中标 准差 $  \sigma $ 已知, 求均值 $  \mu $ 的最大似然估计。</p>
$$
\begin{array}{1}
\theta =\mu \Rightarrow \hat{\theta} =\underset{\theta}{\arg \max } \sum_{k=1}^{N} \log \left(x_{k} \mid \theta\right) \\
 =\underset{\theta}{\arg \max } \sum_{k=1}^{N} \log \left(\frac{1}{\sqrt{\pi} \sigma} \exp \left(-\frac{1}{2 \sigma^{2}}\left(x_{k}-\mu\right)^{2}\right)\right) \\
 =\underset{\theta}{\arg \max } \sum_{k=1}^{N}\left\{\log \frac{1}{\sqrt{\pi} \sigma}-\frac{1}{2 \sigma^{2}}\left(x_{k}-\mu\right)^{2}\right\} \\
\frac{\partial}{\partial \mu} \sum_{k=1}^{N}\left\{\log \frac{1}{\sqrt{\pi} \sigma}-\frac{1}{2 \sigma^{2}}\left(x_{k}-\mu\right)^{2}\right\}=0 \Rightarrow \hat{\theta}=\mu=\frac{1}{N} \sum_{k=1}^{N} x_{k}
\end{array}
$$<p>$ \hat{\theta}$ 就是数据集的样本均值</p>

<h3 id="贝叶斯估计-(Bayes)">贝叶斯估计 (Bayes)<a class="anchor-link" href="#贝叶斯估计-(Bayes)">¶</a></h3><p>已知: 数据集 $ \mathbf{X}=\left\{\mathbf{x}_{1}, \mathbf{x}_{2}, \ldots, \mathbf{x}_{N}\right\} $ 和参数 $ \theta $ 的先验概率 $ p(\theta)$ 
求解: $ p(\mathbf{x} \mid \mathbf{X})$ 
$$
\begin{aligned}
p(\mathbf{x} \mid \mathbf{X}) &amp;=\int p(\mathbf{x} \mid \theta) p(\theta \mid \mathbf{X}) d \theta \\
p(\theta \mid \mathbf{X}) &amp;=\frac{p(\mathbf{X} \mid \theta) p(\theta)}{p(\mathbf{X})}=\frac{p(\mathbf{X} \mid \theta) p(\theta)}{\int p(\mathbf{X} \mid \theta) p(\theta) d \theta} \\
p(\mathbf{x} \mid \theta, \mathbf{X}) &amp;=\prod_{k}^{N} p\left(\mathbf{x}_{k} \mid \theta\right)
\end{aligned}
$$
由于后验分布是一个条件分布, 通常我们取后验分布的期望 $ {\hat\theta_{\text {Bayes }}}=E[\theta \mid \mathbf{X}] $ 作为参数的估计值。</p>



<h2 id="非参数估计-—-Lazy-Learning">非参数估计 — Lazy Learning<a class="anchor-link" href="#非参数估计-—-Lazy-Learning">¶</a></h2><ul>
<li>在非参数估计中，假定相似的输⼊具有相似的输出，不对基础密度假定任何形式的先验参数</li>
<li>非参数模型的复杂性依赖于训练集的⼤⼩，依赖于数据中问题的固有复杂性</li>
<li>当给定训练集时，并不计算模型，而将模型的计算推迟到给定一个检验实例时才进⾏，这会导致存储和计算量的增加。(比如：开卷考试)</li>
<li>核心思路：⼀个向量 落⼊区间 $R$ 中的概率为 $ p=\int_{R} p(x) d x $</li>
<li>主要方法：直方图估计、核估计、k近邻估计
<img alt="image.png" src="https://s2.loli.net/2022/10/20/5vkSjX9TRCdsEKp.png" style="zoom:100%"/></li>
<li>当样本数 $ \mathrm{n} $ 足够大时, 可以近似地认为 $ P \approx \frac{k}{n} $ , 其中 $ \mathrm{k} $ 是出现该特征的频数。</li>
<li>假设密度函数 $ p(x) $ 是连续的, 那么在区域 $ R $ 足够小时, 我们可以近似地认为 $ p(x) $ 是一个常值函数, 因此 $ P \approx p(x) V $, 其中  $V $ 是区域 $ R $ 的体积。
$$ \frac{k}{n} \approx P \approx p(x) V \Rightarrow p(x) \cong \frac{k}{n V}$$
<img src="https://s2.loli.net/2022/10/20/eiRhoIztdyHNXuk.png" style="zoom:50%"/></li>
</ul>

```python

import numpy as np
import sys
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

```

```python

# 生成正态分布数据
def Generate_Sample_Gaussian(mean, cov, P, label):
    '''
        mean 为均值向量
        cov 为方差矩阵a
        P 为单个类的先验概率
        return 单个类的数据集
    '''
    temp_num = round(1200 * P)
    x, y = np.random.multivariate_normal(mean, cov, temp_num).T
    z = np.ones(temp_num) * label
    X = np.array([x, y, z])
    return X.T

```

```python

def Generate_DataSet_plot(mean, cov, P):
    # 画出不同先验对应的散点图
    xx = []
    label = 1
    for i in range(3):
        xx.append(Generate_Sample_Gaussian(mean[i], cov, P[i], label))
        label += 1
        i = i + 1
    # 画图
    plt.figure()
    for i in range(3):
        plt.plot(xx[i][:, 0], xx[i][:, 1], '.', markersize=4.)
        plt.plot(mean[i][0], mean[i][1], 'r*')
    plt.show()
    return xx

```

```python

mean = np.array([[1, 4], [4, 1], [8, 4]])  # 均值数组
cov = [[2, 0], [0, 2]]  # 方差矩阵
num = 1200  # 样本个数
P1 = [1 / 3, 1 / 3, 1 / 3]  # 样本X1的先验概率
P2 = [0.6, 0.1, 0.3]  # 样本X2的先验概率
X1 = np.array(Generate_DataSet_plot(mean, cov, P1), dtype=object)
X2 = np.array(Generate_DataSet_plot(mean, cov, P2), dtype=object)
X1 = np.vstack(X1)
X2 = np.vstack(X2)

```

```python
# 数据集进行分类，不带标签
def classify_dataset_droplabel(X):
    flag1=0
    flag2=0
    flag3=0
    num=X.shape[0]
    label1_X=np.empty(shape=(0,3))
    label2_X=np.empty(shape=(0,3))
    label3_X=np.empty(shape=(0,3))
    for i in range(num):
        if X[i][2]==1:
            if flag1:
                label1_X=np.vstack((label1_X,X[i][0:2]))
            else:
                label1_X=X[i][0:2]
                flag1=1
           
        elif X[i][2]==2:
            if flag2:
                label2_X=np.vstack((label2_X,X[i][0:2]))
            else:
                label2_X=X[i][0:2]
                flag2=1
        elif X[i][2]==3:
            if flag3:
                label3_X=np.vstack((label3_X,X[i][0:2]))
            else:
                label3_X=X[i][0:2]
                flag3=1
    return label1_X,label2_X,label3_X
    
```

```python
# 数据集进行分类，带标签
def classify_dataset(X):
    flag1=0
    flag2=0
    flag3=0
    num=X.shape[0]
    label1_X=np.empty(shape=(0,3))
    label2_X=np.empty(shape=(0,3))
    label3_X=np.empty(shape=(0,3))
    for i in range(num):
        if X[i][2]==1:
            if flag1:
                label1_X=np.vstack((label1_X,X[i]))
            else:
                label1_X=X[i]
                flag1=1
           
        elif X[i][2]==2:
            if flag2:
                label2_X=np.vstack((label2_X,X[i]))
            else:
                label2_X=X[i]
                flag2=1
        elif X[i][2]==3:
            if flag3:
                label3_X=np.vstack((label3_X,X[i]))
            else:
                label3_X=X[i]
                flag3=1
    return label1_X,label2_X,label3_X
    
```

```python
def mu_cov(X):
    mu = np.mean(X, axis=0)
    # python把向量转化成矩阵需要用reshape
    cov = np.array([np.dot((X[i] - mu).reshape(2,1), (X[i] - mu).reshape(1, 2)) for i in range(len(X))]).mean(axis=0)
    # print(mu,cov)
    return mu,cov
    
```

```python
# 极大似然估计
# 输入n*2维数据
def LikelyHood(X):
    # 这里可以修改一下如何得到labelX
    label1_X,label2_X,label3_X=classify_dataset_droplabel(X)
    
    mu1,cov1=mu_cov(label1_X)
    mu2,cov2=mu_cov(label2_X)
    mu3,cov3=mu_cov(label3_X)
    mu=np.vstack((mu1,mu2,mu3))
    cov=np.vstack((cov1,cov2,cov3))

    return mu, cov

```

```python
mu_x1,cov_X1=LikelyHood(X1)
mu_x2,cov_X2=LikelyHood(X2)
```

```python
# Hint for 初级要求：二元高斯分布概率密度函数计算
# 在公式中，x和mean应该是列向量，但是为了方便，这里接收的都是行向量（维度：1*2）
def Gaussian_function(x, mu, cov):
    # 需要先变化cov的类型dtype=float64
    cov=cov.astype('float64')
    det_cov = np.linalg.det(cov)  # 计算方差矩阵的行列式
    inv_cov = np.linalg.inv(cov)  # 计算方差矩阵的逆
    # 计算概率p(x|w)
    p = 1 / (2 * np.pi * np.sqrt(det_cov)) * np.exp(-0.5 * np.dot(np.dot((x - mu), inv_cov), (x - mu).T))
    return p

```

```python
# 似然率测试规则
def Likelihood_Test_Rule(X, mean, cov):
    class_num = mean.shape[0]  # 类的个数
    num = np.array(X).shape[0]
    error_rate = 0
    for i in range(num):
        p_temp = np.zeros(3)
        for j in range(class_num):
        	# 计算样本i决策到j类的概率
            p_temp[j] = Gaussian_function(X[i][0:2], mean[j], cov[2*j:2*j+2])  
        p_class = np.argmax(p_temp) + 1  # 得到样本i决策到的类
        if p_class != X[i][2]:
            error_rate += 1
    return round(error_rate / num , 3)
```

```python
# 最大后验测试规则            P是先验概率
def MaxLast_Test_Rule(X, mean, cov,P):
    class_num = mean.shape[0]  # 类的个数
    num = np.array(X).shape[0]
    error_rate = 0
    for i in range(num):
        p_temp = np.zeros(3)
        for j in range(class_num):
        	# 计算样本i决策到j类的概率
            p_temp[j] = Gaussian_function(X[i][0:2], mean[j], cov[2*j:2*j+2])*P[j]  
        p_class = np.argmax(p_temp) + 1  # 得到样本i决策到的类
        if p_class != X[i][2]:
            error_rate += 1
    return round(error_rate / num , 3)
```

```python
# 单次试验求不同准则下的分类误差
def repeated_trials(mean1, cov1,mean2,cov2, P1, P2, X1, X2):
    # 通过不同规则得到不同分类错误率并返回
    error = np.zeros((2, 2))
    # 计算似然率测试规则误差
    error_likelihood = Likelihood_Test_Rule(X1, mean1, cov1)
    error_likelihood_2 = Likelihood_Test_Rule(X2, mean2, cov2)
    error[0] = [error_likelihood, error_likelihood_2]
    # 计算最大后验概率规则误差
    error_MaxLast_Test_Rule = MaxLast_Test_Rule(X1, mean1, cov1, P1)
    error_MaxLast_Test_Rule_2 = MaxLast_Test_Rule(X2, mean2, cov2, P2)
    error[1] = [error_MaxLast_Test_Rule, error_MaxLast_Test_Rule_2]
    return error
```

```python
error_ave = repeated_trials(mu_x1,cov_X1,mu_x2,cov_X2,P1,P2,X1,X2)
print("误差  :  极似然规则   最大后验规则")
print("X1误差:  \t{}   \t{}".format(error_ave[0][0],error_ave[1][0]))
print("X2误差:  \t{}   \t{}".format(error_ave[0][1],error_ave[1][1]))
```

```python
# Hint for 初级要求：高斯核概率密度函数计算
# 在公式中，x和mean应该是列向量，但是为了方便，这里接收的都是行向量（维度：1*2）
# x是对应X里的每一行
def Gaussian_Kernel(x, X, h=2):
    # 计算概率p(x|w)
    p = (1 / (np.sqrt(2 * np.pi) * h)) * np.array([np.exp(-0.5 * np.dot(x - X[i][0:2], x - X[i][0:2]) / (h**2)) for i in range(len(X))]).mean()
    return p

```

```python
# 高斯核函数测试规则            
def Guassker_Test_Rule(X,h):
    label1_X,label2_X,label3_X=classify_dataset_droplabel(X)
    class_num = mean.shape[0]  # 类的个数
    num = np.array(X).shape[0]
    error_rate = 0
    for i in range(num):
        p_temp = np.zeros(3)
        for j in range(class_num):
            if j==0:    
        	# 计算样本i决策到j类的概率
                p_temp[j] = Gaussian_Kernel(X[i][0:2],label1_X, h)
            if j==1:    
        	# 计算样本i决策到j类的概率
                p_temp[j] = Gaussian_Kernel(X[i][0:2],label2_X, h)
            if j==2:    
        	# 计算样本i决策到j类的概率
                p_temp[j] = Gaussian_Kernel(X[i][0:2],label3_X, h)
        p_class = np.argmax(p_temp) + 1  # 得到样本i决策到的类
        if p_class != X[i][2]:
            error_rate += 1
    return round(error_rate / num , 3)
```

```python
err1=Guassker_Test_Rule(X1,1)
err2=Guassker_Test_Rule(X2,1)
print("误差  :  核函数")
print("X1误差:  \t{}  ".format(err1))
print("X2误差:  \t{}  ".format(err2))
```

```python
# 分层抽样，进行数据集分割
def sample_split(X,N):
    # X 数据集，N，N折交叉验证
    X1,X2,X3=classify_dataset(X)
    step1=int(X1.shape[0]/N)
    step2=int(X2.shape[0]/N)
    step3=int(X3.shape[0]/N)
    train_data=[]
    test_data=[]
    for i in range(N):
        if i==N-1:
            temp2=X
            temp1=np.vstack((X1[i*step1:],X2[i*step2:],X3[i*step3:]))
            # print(temp1)
            test_data.append(temp1)
            for i in temp1:
                index = np.where(temp2==i)[0][0]
                temp2=np.delete(temp2,index,axis=0)
                # temp2=np.delete(temp2,)
            train_data.append(temp2)
            break
        temp1=np.vstack((X1[i*step1:(i+1)*step1],X2[i*step2:(i+1)*step2],X3[i*step3:(i+1)*step3]))
        temp2=X
        test_data.append(temp1)
        for i in temp1:
            # print(i)
            # print(temp2[0])
            index = np.where(temp2==i)[0][0]
            temp2=np.delete(temp2,index,axis=0)
                # temp2=np.delete(temp2,)
        train_data.append(temp2)

    return train_data,test_data  
```

```python
# 中级要求 ，进行h的选择          
def h_best(X,N,h):
    train_data,test_data=sample_split(X,N)
    error_rate = [0,0,0,0,0]
    for m in range(len(h)):
        for i in range(N):
            now_train=train_data[i]
            now_test=test_data[i]
            label1_X,label2_X,label3_X=classify_dataset_droplabel(now_train)
            # print(now_test)
            num=now_test.shape[0]
        
            for i in range(num):
                # print(i)
                p_temp = np.zeros(3)
                for j in range(3):
                    if j==0:    
                    # 计算样本i决策到j类的概率
                        p_temp[j] = Gaussian_Kernel(now_test[i][0:2],label1_X, h[m])
                    if j==1:    
                    # 计算样本i决策到j类的概率
                        p_temp[j] = Gaussian_Kernel(now_test[i][0:2],label2_X, h[m])
                    if j==2:    
                    # 计算样本i决策到j类的概率
                        p_temp[j] = Gaussian_Kernel(now_test[i][0:2],label3_X, h[m])
                p_class = np.argmax(p_temp) + 1  # 得到样本i决策到的类
                # print("p_class ",p_class)
                # print("now_test ",now_test[i][2])
                if p_class != now_test[i][2]:
                    # print(error_rate)
                    error_rate[m] += 1
            error_rate[m]=round(error_rate[m]/now_test.shape[0],3)
    print(error_rate)
    return h[np.argmin(error_rate)] 
    
```

### 高级要求
1. 任选一个数据集，在该数据集上应用k-近邻概率密度估计，任选3个k值输出概率密度分布图。


<h3 id="绘制概率密度分布图示例">绘制概率密度分布图示例<a class="anchor-link" href="#绘制概率密度分布图示例">¶</a></h3>


```python
def Kneibor_Eval(X, k):
    num = len(X)
    Xtrain = np.array(X)
    # 生成200*200=40000个采样点，每个采样点对应三类的不同概率
    p = np.zeros((200, 200, 3))
    # 在[-5,15]的范围内，以0.1为步长估计概率密度
    label1_X,label2_X,label3_X=classify_dataset(X)
    for i in np.arange(0, 200):
        for j in np.arange(0, 200):
            '''
            # 生成标准差距离
            # 根据第k个数据点的位置计算V
            # 找到前k个数据点的类别,分别加到对应类的权重上
            # 计算每个采样点的概率密度函数
            
            '''
            x=[-5+0.1*i,-5+0.1*j]
            t = k / num
            for w_class in range(3):
                dis = []
                if w_class==0:
                    for item in label1_X:
                        dis.append(np.sqrt((x[0]-item[0])**2+(x[1]-item[1])**2))
                if w_class==1:
                    for item in label2_X:
                        dis.append(np.sqrt((x[0]-item[0])**2+(x[1]-item[1])**2))
                if w_class==2:
                    for item in label3_X:
                        dis.append(np.sqrt((x[0]-item[0])**2+(x[1]-item[1])**2))
                dis.sort()
                v = (dis[k-1] * np.pi) ** 2    
                p[i][j][w_class] = t / v 


    return p

```
